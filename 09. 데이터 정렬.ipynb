```python
import pandas as pd
df=pd.read_csv("C:/Users/sungb/ML 강의자료/210112_pandas/iris.csv")
df
```

## 특정 변수를 기준으로 정렬

```python
df.sort_values('sepal length (cm)') # ()에 들어간 value를 기준으로 오름차순 정렬
```

```python
df.sort_values('sepal length (cm)',ascending=False)   # ()에 들어간 value를 기준으로 내림차순 정렬
```

## 다수의 변수를 기준으로 정렬

```python
# 앞에 나오는 변수 순으로 나열하고 동률에서 그 다음 변수에 따른 정렬
# 내림차순

df.sort_values(['label','sepal length (cm)'],ascending=False)
```

```python
#오름차순
df.sort_values(['label','sepal length (cm)'])
```

### 변수 별로 다른 방향으로 나열하고 싶은 경우

```python
df.sort_values(['label','sepal length (cm)'],ascending=[True,False])  #label은 오름차순, sepal은 내림차순
```
