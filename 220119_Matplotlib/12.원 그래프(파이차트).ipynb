# 12. 원그래프(파이차트)
원그래프는 동그란 모양에 비율을 나타내기 위해서 주로 사용한다.

```python
import matplotlib.pyplot as plt

import matplotlib
matplotlib.rcParams['font.family'] = "Malgun Gothic"  # windows
#matplotlib.rcParams['font.family'] = "AppleGothic"    # Mac

# 글자 크기
matplotlib.rcParams['font.size']=15

# 마이너스 깨지는거 방지
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
values=[30,25,20,13,10,2] # 꼭 합이 100에 맞춰질 필요는 없다
labels=['Python','Java','C','C#','R','ETC']
```

```python
# values 만 이용한 파이차트
plt.pie(values)
plt.show()
```

```python
plt.pie(values,labels=labels)
plt.show()
```

### 퍼센트 값 추가

```python
plt.pie(values,labels=labels,autopct='%.1f')
plt.show()
```

```python
values_1=[1,1,1,1,1,1]
```

```python
plt.pie(values_1,labels=labels,autopct='%.1f')
plt.show()
```

```python
# % 추가
plt.pie(values,labels=labels,autopct='%.1f%%')
plt.show()
```

```python
# % 추가
plt.pie(values,labels=labels,autopct='%.1f%%')
plt.show()
```

```python
# 시작점 변경
plt.pie(values,labels=labels,autopct='%.1f%%',startangle=90)
plt.show()
```

```python
# 파이방향 조정
plt.pie(values,labels=labels,autopct='%.1f%%',startangle=90,counterclock=False)
plt.show()
```

```python
# 범례 추가
plt.pie(values,labels=labels)
plt.legend()
plt.show()
```

```python
plt.pie(values,labels=labels)
plt.legend(loc=(1.2,0.3))
plt.show()
```

```python
plt.pie(values,labels=labels)
plt.legend(loc=(1.2,0.3),title="언어별 선호도")
plt.show()
```

```python
plt.pie(values,labels=labels,autopct='%.1f%%')
plt.title("언어")
plt.legend(loc=(1.2,0.3),title="언어별 선호도")
plt.show()
```

# DataFrame에서 부터 그래프


```python
import pandas as pd
df=pd.read_csv("salm_data.csv")
```

```python
df
```

```python
grp=df.groupby('학교')
```

```python
values=[grp.size()['북산고'],grp.size()['능남고']]
labels=['북산고','능남고']

plt.pie(values,labels=labels)
plt.title('소속학교')
plt.show()
```
