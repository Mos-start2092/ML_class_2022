# 4.스타일

```python
import matplotlib.pyplot as plt

import matplotlib
matplotlib.rcParams['font.family'] = "Malgun Gothic"  # windows
#matplotlib.rcParams['font.family'] = "AppleGothic"    # Mac

# 글자 크기
matplotlib.rcParams['font.size']=15

# 마이너스 깨지는거 방지
matplotlib.rcParams['axes.unicode_minus'] = False

x=[1,2,3]
y=[2,4,8]
```

```python
plt.plot(x,y)
```

```python
# line 두께
plt.plot(x,y,linewidth=10)
```

```python
plt.plot(x,y,linewidth=5)
```

## 마커

```python
# 마커
plt.plot(x,y,marker="o")
```

```python
# linestyle -> 선 없앰
plt.plot(x,y,marker="o",linestyle="None")
```

```python
plt.plot(x,y,marker="v")
```

```python
# 마커 사이즈
plt.plot(x,y,marker="v",markersize=10)
```

```python
plt.plot(x,y,marker="X",markersize=10)
```

## 구글링 
matplotlib marker -> 정보가 나온다

```python
# 엣지 컬러
plt.plot(x,y,marker="X",markersize=10,markeredgecolor="red")
```

```python
plt.plot(x,y,marker="X",markersize=15,markeredgecolor="red",markerfacecolor="yellow")
```

## 선 스타일

```python
# 점선
plt.plot(x,y,linestyle=":")
```

```python
plt.plot(x,y,linestyle="--")
```

```python
plt.plot(x,y,linestyle="-.")
```

## 색

```python
plt.plot(x,y,color="pink")
```

```python
plt.plot(x,y,color="b")
```

```python
plt.plot(x,y,color="g")
```

## 포맷

```python
plt.plot(x,y,'ro--') # color,marker style, linestyle
```

```python
plt.plot(x,y,'bv:')
```

```python
# line style 없음
plt.plot(x,y,'go')
```

## 축약어

```python
plt.plot(x,y,marker="o",mfc="red",ms=20,mec="blue",ls=":")
```

## 투명도

```python
plt.plot(x,y,marker="o",mfc="red",ms=20,mec="blue",ls=":",alpha=0.2) # alpah 0~1
```

## 그래프 크기

```python
plt.figure(figsize=(10,5))
plt.plot(x,y)
```

```python
plt.figure(figsize=(5,10))
plt.plot(x,y)
```

## 배경색

```python
plt.figure(facecolor="yellow")
plt.plot(x,y)
```
