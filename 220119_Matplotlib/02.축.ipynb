# 2. 축

```python
import matplotlib.pyplot as plt

import matplotlib
matplotlib.rcParams['font.family'] = "Malgun Gothic"  # windows
#matplotlib.rcParams['font.family'] = "AppleGothic"    # Mac

# 글자 크기
matplotlib.rcParams['font.size']=15

# 마이너스 깨지는거 방지
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
x=[1,2,3]
y=[2,4,8]
```

```python
plt.plot(x,y)
plt.title('꺽은선 그래프')
```

```python
# 데이터에 대한 설명을 축에 넣어준다
plt.plot(x,y)
plt.xlabel('X축')
plt.ylabel('Y축')
```

```python
# 축 이름 색 지정
plt.plot(x,y)
plt.xlabel('X축',color='red')
plt.ylabel('Y축',color='green')
```

```python
# 축 위치 변경
# left, center right  & top,center,bottom
plt.plot(x,y)
plt.xlabel('X축',color='red',loc="right")
plt.ylabel('Y축',color='green',loc="top")
```

```python
# 축 눈금
plt.plot(x,y)
plt.xticks([1,2,3])
plt.yticks([3,6,9])
plt.show()
```
