# 13. 산점도 그래프
(2차원 산점도의 경우) 두가지 변수간의 상관관계를 나타낸다

```python
import matplotlib.pyplot as plt

import matplotlib
matplotlib.rcParams['font.family'] = "Malgun Gothic"  # windows
#matplotlib.rcParams['font.family'] = "AppleGothic"    # Mac

# 글자 크기
matplotlib.rcParams['font.size']=15

# 마이너스 깨지는거 방지
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd
df=pd.read_csv("salm_data.csv")
```

### 학년 변수 추가

```python
df['학년']=[3,3,2,1,1,3,2,2]
df
```

## 산점도 그래프

```python
plt.scatter(df['영어'],df['수학'])
```

```python
plt.scatter(df['영어'],df['수학'])
plt.xlabel("영어 점수")
plt.ylabel("수학 점수")
```

### 점 크기 변동

```python
import numpy as np
sizes=np.random.rand(8)
sizes
```

```python
sizes_1=sizes*1000
```

```python
plt.scatter(df['영어'],df['수학'],s=sizes_1)
plt.xlabel("영어 점수")
plt.ylabel("수학 점수")
```

### 사이즈 같은 경우 주로 그룹화 한 뒤 분산의 크기등을 나타내기 위해 사용한다

### 학년 정보로 사이즈 지정

```python
sizes=df['학년']*500
```

```python
plt.scatter(df['영어'],df['수학'],s=sizes)
plt.xlabel("영어 점수")
plt.ylabel("수학 점수")
```

### 색 변경
구글링 -> cmap

```python
plt.scatter(df['영어'],df['수학'],s=sizes,c=df['학년'])
plt.xlabel("영어 점수")
plt.ylabel("수학 점수")
```

```python
plt.scatter(df['영어'],df['수학'],s=sizes,c=df['학년'],cmap='Greys')
plt.xlabel("영어 점수")
plt.ylabel("수학 점수")
```

```python
plt.scatter(df['영어'],df['수학'],s=sizes,c=df['학년'],cmap='winter')
plt.xlabel("영어 점수")
plt.ylabel("수학 점수")
```

```python
plt.scatter(df['영어'],df['수학'],s=sizes,c=df['학년'],cmap='winter',alpha=0.3)
plt.xlabel("영어 점수")
plt.ylabel("수학 점수")
```

```python
plt.figure(figsize=(7,7))
plt.scatter(df['영어'],df['수학'],s=sizes,c=df['학년'],cmap='winter',alpha=0.3)
plt.xlabel("영어 점수")
plt.ylabel("수학 점수")
plt.colorbar()
```

```python
plt.figure(figsize=(7,7))
plt.scatter(df['영어'],df['수학'],s=sizes,c=df['학년'],cmap='winter',alpha=0.3)
plt.xlabel("영어 점수")
plt.ylabel("수학 점수")
plt.colorbar(ticks=[1,2,3],label='학년',shrink=0.5)
```
