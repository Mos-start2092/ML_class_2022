# 데이터 마이닝

- 데이터 마이닝(Data Mining)은 대량의 데이터로부터 규칙이나 패턴을 찾아내는 과정으로, 통계학, 데이터베이스, 기계학습, 인공지능의 영역에서의 발전된 다양한 기법들을 포함


- 데이터 마이닝의 목적은 데이터셋(Data set)으로부터 정보를 추출하고, 이를 추후 사용을 위해 이해할 수 있는 구조로 변환하는 것


- 데이터 마이닝은 크게 EDA(Explorary Data Analysis) -> 전처리 -> 분석 및 모델링 -> 예측 의 과정으로 이루어진다.


- 분석 및 모델링의 과정만을 데이터 마이닝으로 부르기도 한다.



### 데이터 마이닝 적용사례
- 기업의 표적 마케팅
- 고객 세분화 -> 유튜브 알고리즘 
- 금융분야에서의 신용평가
- 제조업에서의 품질관리
- 텍스트 마이닝을 통한 정보검색과 음성과 영상 등의 멀티미디어 자료의 분석


# 머신러닝(기계학습)

- 인공지능의 한 분야로서 컴퓨터로 하여금 센서나 데이터베이스로부터의 자료를 이용하여 판단하도록 하는 알고리즘을 개발하는 학문분야


- 자료에 기반하여 복잡한 패턴을 인식하고 의사결정을 내릴 수 있게 자동적으로 학습하는데에 중점 


- 넓게는 분석과정에서 기계학습(Machine Learning)을 활용하는 데이터 마이닝을 머신러닝이라고 부른다. 




# 머신러닝의 과정 요약

### 1. 훈련용 데이터(train data)로부터 패턴을 발견하여 모델을 학습한다
- 현재 소유중인, 즉 과거의 데이터인 훈련용 데이터(train data)를 분석한다
- EDA(탐색적 데이터 분석), 데이터 전처리, 모델링 및 평가 과정을 통해 분석하고 패턴을 발견한다

### 예시


- X1: 몸무게
- X2: 키 
- 빨간 데이터 = 비만  / 초록 데이터 =  정상
- 파란색 라인 = 발견해낸 패턴(함수) = 학습된 모델

### 2.테스트 데이터(test data)가 패턴을 기준으로 어디에 속하는지 판단하고 비만인지 아닌지 판단한다
- 분석 or 예측을 해야하는 새로운 데이터를 훈련된 패턴에 적용한 뒤 결과를 해석한다


# 지도학습 vs 비지도학습
 머신러닝(기계학습)은 크게 지도학습(supervised learning)과 비지도학습(unsupervised learning)으로 나뉜다.



## 지도학습(supervised learning) 
- 결과값이 알려진 상황에서의 학습모형을 활용한 예측 모형


-  목표마케팅, 성과예측, 의학진단, 사기검출 등의 분야에서 사용


-  크게 r결과값이 연속인 데이터를 예측하는 회귀와 결과값이 범주형인 데이터를 예측하는 분류로 나뉨 


#### 회귀 예시
주식보유기간 예측, 집값 예측, 태양광 발전량 예측

####  분류 예시
이미지 분류, 파산여부 예측, 암 예측, 이탈고객 예측

## 비지도학습(unsupervised learning)
- 별도의 결과값을 요구하지 않는 자료에 대한 분석

# 데이터 구조 보는 법

- 지도학습의 경우 알고 싶은 값을 target으로 잡고 분석이 실시된다. 

- 이때 target이 되는 변수를 예측변수(종속변수), 나머지 변수를 설명변수(독립변수)로 표현한다. 


- 지도학습에서 데이터의 패턴을 알고 싶다는 것은 예측변수를 설명변수들로 나타내는 함수를 찾고 싶다는 것으로 해석할 수 있다.


- 그러므로 설명변수를 Y, 설명변수들을 X1, X2,,, 와 같이 X로 생각하고 다음과 같은 함수를 찾는 것으로 해석할 수도 있다. 




- 위의 함수에서 앱실론으로 표현된 오차(error)는 현실관계성 이외에도 외부로 부터의 영향으로 결과를(Y) 다르게 하는 효과를 나타낸다

#### 3. 변수 선정

#### 위의 EDA와 전처리를 기반으로 변수선정 및 변수변환을 실시

```python
# 목표 변수 분리 - 출력변수
df_churn = df['Churn']

# 모델 학습에 이용할 변수 선정 - 입력변수
df_X = df[['InternetService','OnlineBackup','TechSupport','tenure','MonthlyCharges','TotalCharges']]

df_X.head()
```

#### 4. 범주형 변수 더미 변수화

```python
# 더미변수화
df_X = pd.get_dummies(df_X)

# 목표 변수(Churn) 과 결합
df_new = pd.concat([df_X,df_churn], axis = 1)

df_new.head()
```

```python
# 다시 train , test 분리

train_new=df_new.iloc[:2000,:]
test_new=df_new.iloc[2001:,:]
```

## 모델 학습 및 평가

```python
X_train=train_new.drop(['Churn'],axis=1)
y_train=train_new['Churn']
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
X_train.info()
```

### 의사결정 나무를 활용한 모델학습

```python
# 모델 학습
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(max_depth = 6)

model = model.fit(X_train, y_train)
```

```python
model
```

### 학습된 모델을 활용한 예측

```python
X_test=test_new.drop(['Churn'],axis=1)

pred = model.predict(X_test)
```

예측결과값 확인

```python
pred
```

### 실제 test_y 값과 비교하여 정확도 평가

```python
from sklearn.metrics import accuracy_score

y_test=test_new['Churn']


print('Accuracy_score: %.3f' % accuracy_score(y_test, pred))
```

```python
# ROC curve
from sklearn.metrics import roc_curve

def plot_roc_curve(fper, tper):
    plt.plot(fper, tper, color='red', label='ROC')
    plt.plot([0, 1], [0, 1], color='green', linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic Curve')
    plt.legend()
    plt.show()
    

prob = model.predict_proba(X_test)
prob = prob[:, 1]
fper, tper, thresholds = roc_curve(y_test, prob)
plot_roc_curve(fper, tper)
```
