# 4. 스타일

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = "Malgun Gothic"

# 글자 크기
matplotlib.rcParams['font.size']=15

# 마이너스 
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
x=[1,2,3]
y=[2,4,8]
```

```python
plt.plot(x,y)
```

```python
# line 두께
plt.plot(x,y,linewidth=10)
```

```python
plt.plot(x,y,linewidth=5)
```

## 마커

```python
plt.plot(x,y,marker="o")
```

```python
plt.plot(x,y,marker="o",linestyle="None")
```

```python
plt.plot(x,y,marker="v",markersize=10)
```

```python
plt.plot(x,y,marker="X",markersize=10)
```

```python
plt.plot(x,y,marker="s",markersize=10)
```

```python
# 엣지 컬러
plt.plot(x,y,marker="X",markersize=10,markeredgecolor="red")
```

```python
plt.plot(x,y,marker="X",markersize=10,markeredgecolor="red",markerfacecolor="yellow")
```

## 선 스타일

```python
# 점선 
plt.plot(x,y,linestyle=":")
```

```python
plt.plot(x,y,color="b")
```

```python
plt.plot(x,y,color="g")
```

```python
plt.plot(x,y,marker="X",ms=10,mec="red",mfc="yellow",ls=":")
```

```python
plt.figure(figsize=(7,7))
plt.plot(x,y,marker="X",ms=10,mec="red",mfc="yellow",ls=":")
```

## 배경색


```python
plt.figure(facecolor="yellow")
plt.plot(x,y)
```
