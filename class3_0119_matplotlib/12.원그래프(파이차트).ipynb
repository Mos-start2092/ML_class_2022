# 12.원그래프(파이차트)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = "Malgun Gothic"

# 글자 크기
matplotlib.rcParams['font.size']=15

# 마이너스 
matplotlib.rcParams['axes.unicode_minus']=False


import pandas as pd
df=pd.read_csv("salm_data.csv")
```

```python
df
```

```python
values=[30,25,20,13,10,2]
label_1=['Python','Java','C','C#','R','ETC']
```

```python
# 파이차트
plt.pie(values)
plt.show()
```

```python
plt.pie(values,labels=label_1)
plt.show()
```

# 퍼센트 추가

```python
plt.pie(values,labels=label_1,autopct='%.1f')
plt.show()
```

```python
values_1=[1,1,1,1,1,1]

plt.pie(values_1,labels=label_1,autopct='%.1f')
plt.show()
```

```python
# % 추가
plt.pie(values,labels=label_1,autopct='%.1f%%')
plt.show()
```

```python
# 시작점 변경
plt.pie(values,labels=label_1,autopct='%.1f%%',startangle=90)
plt.show()
```

```python
# 파이방향 조정
plt.pie(values,labels=label_1,autopct='%.1f%%',startangle=90,counterclock=False)
plt.show()
```

```python
# 범례추가
plt.pie(values,labels=label_1,autopct='%.1f%%',startangle=90,counterclock=False)
plt.legend()
plt.show()
```

```python
plt.pie(values,labels=label_1,autopct='%.1f%%',startangle=90,counterclock=False)
plt.legend(loc=(1.2,0.3))
plt.show()
```

```python
plt.pie(values,labels=label_1,autopct='%.1f%%',startangle=90,counterclock=False)
plt.title("언어")
plt.legend(loc=(1.2,0.3),title="언어별 선호도")
plt.show()
```

## 데이터 프레임에서 그래프

```python
import pandas as pd
df=pd.read_csv("salm_data.csv")
```

```python
df
```

```python
grp=df.groupby('학교')
```

```python
values=[grp.size()['북산고'],grp.size()['능남고']]
labels=['북산고','능남고']

plt.pie(values,labels=labels,autopct='%.1f%%')
plt.title('학교')
plt.show()
```
