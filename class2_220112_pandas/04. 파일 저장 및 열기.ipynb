# 4. 파일 저장 및 열기
DataFrame 객체를 excel, csv 형태의 파일로 저장 및 열기
(txt의 형태 파일은 거의 사용하지 않기에 생략)

```python
import pandas as pd


df_temp=pd.DataFrame({
    'temp_low' :[-25,-15,5,15],
    'temp_mean':[-20,-10,10,20],
    'temp_high':[-15,-5,15,25]},
    index=["Jan","Feb","Mar","Apr"])
df_temp.index.name="월"
df_temp
```

## 저장하기
### csv 파일로 저장
csv 파일 - 쉼표로 나눠진 텍스트 형태의 파일이며 엑셀에서 부르면 표로 나눠서 나온다

```python
df_temp.to_csv("temp.csv")
```

```python
# 엑셀에서 확인해보면 깨진 것을 볼 수 있다. 이는 인코딩 문제
# option으로 해결
df_temp.to_csv("temp.csv",encoding="utf-8-sig")
```

```python
# index 제거
df_temp.to_csv("temp.csv",encoding="utf-8-sig",index=False)
```

#### 귀찮은 과정을 생략하기 위해 한글을 포함하지 않도록 데이터 생성하고 진행

```python
df_temp=pd.DataFrame({
    'temp_low' :[-25,-15,5,15],
    'temp_mean':[-20,-10,10,20],
    'temp_high':[-15,-5,15,25]},
    index=["Jan","Feb","Mar","Apr"])
df_temp.index.name="month"
df_temp
```

### 엑셀 파일로 저장


```python
df_temp.to_excel('temp.xlsx')
```

## 불러오기
