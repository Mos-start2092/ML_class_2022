# 11. 그룹화
동일한 값을 가진 것들끼리 합쳐서 통계 또는 평균 등의 값을 계산하기 위해 사용

```python
import pandas as pd
df=pd.read_csv("iris.csv")
```

```python
df
```

```python
df.groupby("label") 
```

```python
# 특정 그룹만 가져옴
df.groupby("label").get_group(2)
```

```python
df.groupby("label").get_group(1)
```

```python
df.groupby('label').mean() # 그룹에 따라 계산 가능한 데이터들의 평균값
```

```python
df.groupby('label').size()  # 각 그룹의 크기 정보
```

```python
df.groupby('label').size()[0]  # 그룹화한 뒤에 0그룹의 size만 본다
```

```python
df.groupby('label')['sepal length (cm)'].mean()   # label로 그룹화한 뒤에 특정 변수에 대해서만 값을 알고 싶음
```

```python
df.groupby('label')[['sepal length (cm)','sepal width (cm)']].mean() # 그룹화한 뒤 다변수에 대해 정보 추출
```

### 두번의 그룹화

```python
df['sepal']=df['sepal length (cm)'].round()  # 반올림
df
```

```python
df['sepal'].unique()
```

```python
df.groupby(['label','sepal']).mean() # label 별, sepal 별 그룹화 해서 평균을 보여줌
```

```python
df.groupby(['sepal','label']).mean()
```

```python
# label로 groupby를 하고 정렬 가능 
df.groupby('label').mean().sort_values('sepal width (cm)')
```

```python
# 합
df.groupby('label').sum()
```

```python
# 표준편차
df.groupby('label').std()
```

```python
# count
df.groupby('label').count()
```

```python
# nunique
df.groupby('label').nunique()
```

```python
label=df.groupby('label')
label['sepal'].value_counts() # label로 그룹화 한뒤에 sepal의 수를 가져온것
```
