# 8. 데이터 선택(조건)
- < : 미만, 초과
- <= : 이하, 이상
- == : 동일한가
- != : 다른가

```python
import pandas as pd
df=pd.read_csv("C:/Users/sungb/ML 강의자료/210112_pandas/iris.csv")
df
```

```python
# 조건에 부합하는지 살펴본다
df['sepal length (cm)'] >= 7   # sepal length가 7이상인지 True, False 보여줌
```

```python
# 조건에 맞는 데이터 출력
df[df['sepal length (cm)'] >= 7 ] # True인 인덱스만 출력
```

```python
# 조건에 맞는 특정 컬럼의 데이터만 추출
df[df['sepal length (cm)'] >= 7 ].iloc[:,[3,4]]  # 조건이 맞는 데이터 중에서 3,4 열만 출력
```

## 다양한 조건

### & 그리고

```python
df[(df['sepal length (cm)'] >= 7) & (df['label']==2)] # sepal이 7이상이고 label이 2인 데이터만 추출
```

### \ 또는 (or)
Enter위에 달러표시


```python
df[(df['sepal length (cm)'] >= 7) | (df['sepal width (cm)']==2.5)]
```
