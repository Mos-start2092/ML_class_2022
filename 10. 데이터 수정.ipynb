# 10. 데이터 수정

```python
import pandas as pd
df=pd.read_csv("C:/Users/sungb/ML 강의자료/210112_pandas/iris.csv")
df
```

## column 수정 

```python
# label을 0,1,2가 아니라 1,2,3으로 만들고 싶다
# 0을 3으로 변환해주자
df['label'].replace({0:3})
```

```python
# inplace해주지 않으면 그대로다
df
```

```python
df['label'].replace({0:3},inplace=True)
```

```python
df
```

## Column 추가

```python
df['sepal sum']=df['sepal length (cm)']+df['sepal width (cm)']
```

```python
df
```

```python
# sepal length가 6m 이상인 애들을 나타내는 열을 추가하고 싶음

df['sepal long']= df['sepal length (cm)']>=6
```

```python
df
```

## column 삭제

```python
df.drop(columns=['sepal long']) # sepal long 컬럼 삭제
```

```python
df.drop(columns=['sepal long','sepal sum'])
```

- inplace해주면 기존의 데이터도 변경된다

## Row 삭제

```python
df.drop(index=3) # index=3인 행을 삭제
```
